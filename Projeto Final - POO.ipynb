```python
import sys

import e_farmacia.Cliente as Cliente
import e_farmacia.Laboratorio as Laboratorio
import e_farmacia.e_medicamentos.MedicamentoQuimioterapico as MedicamentoQuimioterapico
import e_farmacia.e_medicamentos.MedicamentoFitoterapico as MedicamentoFitoterapico

def cadastrar_decorator(func):
    def wrapper(*args, **kwargs):
        try:
            return func(*args, **kwargs)
        except Exception as e:
            print("Erro:", e)
            sys.exit()
    return wrapper

def cadastrar_cliente(cpf, nome, data_nasc):
    if len(cpf) != 11:
        print("CPF inválido")
        sys.exit()
    if nome == "" or data_nasc == "":
        print("Nome ou data de nascimento inválido")
        sys.exit()
    if cliente_existe(cpf):
        print("Cliente já cadastrado")
        sys.exit()
    clientes.append(Cliente(cpf, nome, data_nasc))
    print("Cliente cadastrado com sucesso")

def buscar_cliente_por_cpf(cpf):
    for cliente in clientes:
        if cliente.cpf == cpf:
            return cliente
    return None

def cadastrar_medicamento_quimioterapico(nome, principio_ativo, laboratorio, descricao, precisa_receita):
    if nome == "" or principio_ativo == "" or laboratorio == "" or descricao == "" or precisa_receita not in (True, False):
        print("Nome, princípio ativo, laboratório, descrição ou precisa_receita inválido")
        sys.exit()
    if medicamento_existe(nome):
        print("Medicamento já cadastrado")
        sys.exit()
    medicamentos_quim.append(MedicamentoQuimioterapico(nome, principio_ativo, laboratorio, descricao, precisa_receita))
    print("Medicamento quimioterápico cadastrado com sucesso")

def cadastrar_medicamento_fitoterapico(nome, principio_ativo, laboratorio, descricao):
    if nome == "" or principio_ativo == "" or laboratorio == "" or descricao == "":
        print("Nome, princípio ativo, laboratório ou descrição inválido")
        sys.exit()
    if medicamento_existe(nome):
        print("Medicamento já cadastrado")
        sys.exit()
    medicamentos_fit.append(MedicamentoFitoterapico(nome, principio_ativo, laboratorio, descricao))
    print("Medicamento fitoterápico cadastrado com sucesso")

def cadastrar_laboratorio(nome, endereco, telefone, cidade, estado):
    if nome == "" or endereco == "" or telefone == "" or cidade == "" or estado == "":
        print("Nome, endereço, telefone, cidade ou estado inválido")
        sys.exit()
    laboratorios.append(Laboratorio(nome, endereco, telefone, cidade, estado))
    print("Laboratório cadastrado com sucesso")

def medicamento_existe(nome):
    for medicamento in medicamentos_quim + medicamentos_fit:
        if medicamento.nome == nome:
            return True
    return False

def cliente_existe(cpf):
    for cliente in clientes:
        if cliente.cpf == cpf:
            return True
    return False
```

```python
# Exemplo de uso
clientes = []
medicamentos_quim = []
medicamentos_fit = []
laboratorios = []
vendas = []

# Cadastro de Clientes, Medicamentos e Laboratórios

# Menu e interações com o usuário
while True:
    print("Menu:")
    print("1. Cadastrar Cliente")
    print("2. Cadastrar Medicamento Quimioterápico")
    print("3. Cadastrar Medicamento Fitoterápico")
    print("4. Cadastrar Laboratório")
    print("5. Sair")
    
    escolha = input("Escolha uma opção: ")
    
    if escolha == "1":
        cpf = input("CPF do cliente: ")
        nome = input("Nome do cliente: ")
        data_nasc = input("Data de nascimento do cliente: ")
        cadastrar_cliente(cpf, nome, data_nasc)
    elif escolha == "2":
        nome = input("Nome do medicamento: ")
        principio_ativo = input("Princípio ativo: ")
        laboratorio = input("Laboratório: ")
        descricao = input("Descrição: ")
        precisa_receita = input("Precisa de receita (True/False): ")
        cadastrar_medicamento_quimioterapico(nome, principio_ativo, laboratorio, descricao, precisa_receita)
    elif escolha == "3":
        nome = input("Nome do medicamento: ")
        principio_ativo = input("Princípio ativo: ")
        laboratorio = input("Laboratório: ")
        descricao = input("Descrição: ")
        cadastrar_medicamento_fitoterapico(nome, principio_ativo, laboratorio, descricao)
    elif escolha == "4":
        nome = input("Nome do laboratório: ")
        endereco = input("Endereço: ")
        telefone = input("Telefone: ")
        cidade = input("Cidade: ")
        estado = input("Estado: ")
        cadastrar_laboratorio(nome, endereco, telefone, cidade, estado)
    elif escolha == "5":
        print("Saindo...")
        break
    else:
        print("Opção inválida")
```
